```python
train_data=train_data.drop(columns='Unnamed: 0')
test_data=test_data.drop(columns='Unnamed: 0')
```

## Encodeing the dataset

```python
encoder=LabelEncoder()
for i in train_data.columns:
  train_data[i]=encoder.fit_transform(train_data[i])

for i in test_data.columns:
  test_data[i]=encoder.fit_transform(test_data[i])
```

```python
train_data.head()
```

```python
test_data.head()
```

## Spliting the Dataset

```python
x_train=train_data.drop(columns='is_fraud')
y_train=train_data['is_fraud']
x_test=test_data.drop(columns='is_fraud')
y_test=test_data['is_fraud']
```

```python
rd_train=x_train[:2000]
dr_train=y_train[:2000]
```

## Building and Training the Model

Logistic Regression

```python
modelLR=LogisticRegression(max_iter=1000, random_state=42).fit(x_train,y_train)
predictionsLR=modelLR.predict(x_test)
cmLR=confusion_matrix(y_test,predictionsLR)
accuracyLR=accuracy_score(y_test,predictionsLR)
print(cmLR,accuracyLR)
```

Random Forest

```python
modelRF=RandomForestClassifier(random_state=42).fit(rd_train,dr_train)
predictionsRF=modelRF.predict(x_test)
cmRF=confusion_matrix(y_test,predictionsRF)
accuracyRF=accuracy_score(y_test,predictionsRF)
print(cmRF,accuracyRF)
```

Decission Tree

```python
modelDT=DecisionTreeClassifier(random_state=42).fit(x_train,y_train)
predictionsDT=modelDT.predict(x_test)
cmDT=confusion_matrix(y_test,predictionsDT)
accuracyDT=accuracy_score(y_test,predictionsDT)
print(cmDT,accuracyDT)
```
